## One hot encoding to categorial features

```python
from sklearn.preprocessing import LabelEncoder

categorial_features = ['POLICY_BRANCH', 'VEHICLE_MAKE', 'VEHICLE_MODEL', 'POLICY_INTERMEDIARY', 'INSURER_GENDER', 'POLICY_CLM_N', 
                       'POLICY_CLM_GLT_N', 'POLICY_PRV_CLM_N', 'POLICY_PRV_CLM_GLT_N', 'POLICY_YEARS_RENEWED_N', 'CLIENT_REGISTRATION_REGION']
numeric_features = list(set(data.columns) - set(categorial_features) - {'DATA_TYPE'} - {'POLICY_IS_RENEWED'})
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
cat_transform_data = ohe.fit_transform(data[categorial_features]).toarray()
cat_transform_data.shape
```

## Label encoding for categorial features


```python
from sklearn.preprocessing import LabelEncoder
data_le = pd.DataFrame(data=data[categorial_features],columns=categorial_features)

for column in categorial_features:
    le = LabelEncoder()
    data_le[column] = le.fit_transform(data[column])
```

## Scale numeric features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_transform_data = scaler.fit_transform(data[numeric_features])
num_transform_data.shape
```

```python
cat_transform_data.shape, num_transform_data.shape
```

```python
import numpy as np
transformed_data = np.concatenate((cat_transform_data, num_transform_data), axis=1)
transformed_data.shape
```

```python
train_indices = data['DATA_TYPE'] == 'TRAIN'
train = transformed_data[train_indices]
train
y_train = y[train_indices]
```

```python
test_indices = data['DATA_TYPE'] == 'TEST '
test = transformed_data[test_indices]
test
y_test = y[test_indices]
```

```python
len(test) + len(train) == len(data)
```

## Ridge regression

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import confusion_matrix

ridge = RidgeClassifier()
ridge.fit(train, y_train)
tn, fp, fn, tp = confusion_matrix(y_test, ridge.predict(test)).ravel()
```

```python
tn, fp, fn, tp
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(ridge, train, y_train, cv=5,
                         scoring=('accuracy'),
                         return_train_score=True)

scores
```

```python
scores['test_score'].mean()
```
